# 結論

首先進行各區域總車位數計算，以板橋、三重、新莊為最多前三名

接下來進行各站點平均車位進行分析，發現各站點的車位相當平均，約為30~40之間

接下來進行各區域站點密度分析，用人來看，幾個人擁有一個車位，發現人對於Youbike在設置站點與規畫車位數的影響大，有些區人口比較少甚至就沒有youbike站

接下來換帶入面積變數進行計算，密度越高代表該區站與站之間的距離越長，需要增設youbike站，不過這些區域通常是人口較少區域。

最後將所有變數歸納到一張表內，進行相關性分析，其結果的確印證，先前之假設，youbike公司設站點之優先考量為人口變數
